```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Task 1: Load Titanic Dataset
import seaborn as sns
import matplotlib.pyplot as plt
# Load the dataset (ensure Titanic-Dataset.csv is uploaded in the same directory)
url="https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
df = pd.read_csv(url)
df

```

```python
#Task 2: Inspect the data ---
print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())
print("\nData Types:\n", df.dtypes)
print("\nDescriptive Stats:\n", df.describe())
```

```python
#Task 3: Create new categorical features for Pclass and Cabin ---
df['Pclass_cat'] = df['Pclass'].map({1: 'First', 2: 'Second', 3: 'Third'})
```

```python
#Task 4: Handle missing Age values ---
print("Missing Age values:", df['Age'].isnull().sum())

# Fill with median
df['Age'] = df['Age'].fillna(df['Age'].median())

print("Missing Age values after fill:", df['Age'].isnull().sum())
```

```python
# Task 5: Create a Family feature
df['FamilySize'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']

# Print to verify
print(df[['Siblings/Spouses Aboard', 'Parents/Children Aboard', 'FamilySize']].head())
```

```python
#Task 6: Identify highest fare and largest family size ---
highest_fare = df[df['Fare'] == df['Fare'].max()]
largest_family = df[df['FamilySize'] == df['FamilySize'].max()]

print("Passenger(s) with highest fare:\n", highest_fare[['Name', 'Fare']])
print("\nPassenger(s) with largest family:\n", largest_family[['Name', 'FamilySize']])
```

```python
#Task 7: Series and DataFrame manipulations ---
# Series example
fare_series = df['Fare']
print("\nFirst 5 fares:", fare_series.head())

# DataFrame slicing and indexing
print("\nSubset of passengers (rows 10 to 15 and selected columns):")
print(df.loc[10:15, ['Name', 'Sex', 'Fare']])
```

```python
#Task 8: Multi-index DataFrame of survival rates by Sex and Pclass ---
multi_index_df = df.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()
print("\nSurvival rates by Sex and Pclass:\n", multi_index_df)
```

```python
# Histogram of passenger ages
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar plot of survival rates by Pclass ---
plt.figure(figsize=(6, 4))
sns.barplot(x='Pclass', y='Survived', data=df, ci=None, palette='pastel')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
#Heatmap of correlations ---
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pair plot of selected features ---
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='husl')
plt.suptitle("Pair Plot of Key Features", y=1.02)
plt.show()
```

```python
#Survival by Sex and Class ---
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', hue='Pclass', data=df, palette='Set2')
plt.title("Survival Rate by Sex and Class")
plt.show()
```

```python
# Box plot for Age distribution by Survival ---
plt.figure(figsize=(6, 4))
sns.boxplot(x='Survived', y='Age', data=df, palette='Set3')
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
#Violin plot for Fare distribution by Survival ---
plt.figure(figsize=(6, 4))
sns.violinplot(x='Survived', y='Fare', data=df, palette='muted')
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.show()
```
